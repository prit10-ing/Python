```python
import matplotlib.pyplot as plt              # 'pyplot' is core plotting interface
import numpy as np                           # Numpy is used for numerical/data-generation
```

```python
# ---- 2. Simple Line Plot ----------------------------------------------------
# Use a Line Chart when you want to show a trend/change over continuous values (e.g., time, sequence).

x = [1, 2, 3, 4, 5]                 # Independent variable (e.g., days, hours)
y = [2, 5, 6, 2, 10]                # Dependent variable (e.g., sales, growth)

plt.figure(figsize=(5,2))
plt.plot(x, y)                      # Plots a line connecting (x, y) pairs
plt.title("Simple Line Chart")      # Chart title (makes chart readable)
plt.xlabel("X Values")              # Label for X-axis
plt.ylabel("Y Values")              # Label for Y-axis
plt.show()                          # Displays the plot window
```

```python
# ---- 3. Scatter Plot --------------------------------------------------------
# Use a Scatter Plot when you have two variables (x and y) and want to examine correlation
# or the distribution of individual points (e.g., height vs. weight of people).

x = [1, 2, 3, 4, 5]                 # X: First variable (e.g., test score)
y = [2, 4, 5, 3, 6]                 # Y: Second variable (e.g., hours studied)


plt.figure(figsize=(5,2))
plt.scatter(x, y,marker ="x")                   # Shows each (x, y) as a dot (no line)
plt.title("Simple Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
# ---- 4. Bar Chart -----------------------------------------------------------
# Use a Bar Chart to compare values across categories/groups (e.g., sales by region, counts by type).

categories = ['A', 'B', 'C', 'D']   # Categories (e.g., brands, classes)
values = [5, 3, 7, 4]               # Values for each category

plt.figure(figsize=(5,3))
plt.bar(categories, values,)         # Draws a separate bar for each category
plt.title("Bar Chart Example")
plt.xlabel("Category")              # Categorical X-axis
plt.ylabel("Value")                 # Value Y-axis
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.barh(categories, values,color="black")         # Draws a separate bar for each category
plt.title("Bar Chart Example")
plt.xlabel("Category")              # Categorical X-axis
plt.ylabel("Value")                 # Value Y-axis
plt.show()
```

```python
# ---- 5. Histogram -----------------------------------------------------------
# Use a Histogram to see the distribution of a numeric variable (e.g., age distribution in a survey).

data = np.random.randn(10000)         # 100 random data points (Normal distribution)

plt.hist(data, bins=10)             # Shows how many data points fall into 10 intervals
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# ---- 6. Pie Chart -----------------------------------------------------------
# Use a Pie Chart to show parts of a whole (percentages or proportions).
# Avoid if there are more than 5-6 categories; use bar instead.

sizes = [20, 30, 25, 25]            # Size of each part
labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']

plt.pie(sizes, labels=labels, autopct='%1.1f%%') # autopct shows percentage
plt.title("Pie Chart Example")
plt.show()
```

```python
# ---- 7. Customizing Plots ---------------------------------------------------
# Customization helps make plots easy to understand in reports.

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, color='green', linestyle='--', marker='o', label='x squared')
plt.legend()                        # Shows labels (for multiple datasets)
plt.grid(True)                      # Adds a grid for reading values
plt.title("Customized Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
# ---- 8. Saving Plots --------------------------------------------------------
# Save plots to an image file for reports, presentations, or documentation.

plt.bar(categories, values)
plt.title("Save Me!")
plt.savefig('bar_chart.png')           # Saves without showing window (file will be saved in current directory)
plt.close()         
```

```python
# ---- 9. Plotting Multiple Lines ---------------------------------------------
# Multiple lines help compare trends, e.g., two products' sales over time.

x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]                  # First line (e.g., Product A)
y2 = [1, 4, 9, 16, 25]                # Second line (e.g., Product B)

plt.plot(x, y1, label='Linear')
plt.plot(x, y2, label='Squared')
plt.legend()
plt.title("Multiple Lines Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# ---- ADVANCED: Subplots -----------------------------------------------------
# Subplots are useful when you want to show multiple visual comparisons in one figure.

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(6, 4)) # 1 row, 2 plots
axes[0].plot([1,2,3], [1,4,9])           # Line plot on left
axes[0].set_title('Line Plot')
axes[1].bar(['A', 'B', 'C'], [3,7,4])    # Bar chart on right
axes[1].set_title('Bar Chart')
plt.tight_layout()                       # Prevents labels from overlapping
plt.show()
```

```python
# ---- ADVANCED: Annotating Plots ---------------------------------------------
# Annotations are used to highlight important points (e.g., best result).

x = [1, 2, 3, 4, 5]
y = [3, 7, 2, 9, 5]

plt.plot(x, y, marker='o')
plt.title("Plot with Annotation")
plt.annotate('Highest',                   # Text to annotate
             xy=(4,9),                    # Coordinate to point to
             xytext=(3,9.5),              # Text location
             arrowprops=dict(facecolor='red', shrink=0.05)) # Creates the arrow
plt.show()
```

```python
# ---- ADVANCED: Color Maps ---------------------------------------------------
# Colormaps are useful for showing third variable encoded as color, e.g., density or clusters.

x = np.random.rand(50)           # 50 random x positions
y = np.random.rand(50)           # 50 random y positions
colors = np.random.rand(50)      # Random values for color mapping

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()                   # Shows the scale for the third variable (colors)
plt.title("Scatter with Color Map")
plt.show()
```

```python
# ============================================================================
# When to Use Which Chart:
#  - Line Plot: Trends over time or ordered data.
#  - Scatter Plot: Relationship/correlation between 2 continuous variables.
#  - Bar Chart: Compare values across categories/groups.
#  - Histogram: Distribution/frequency of a single numeric variable.
#  - Pie Chart: Proportional data (parts of a whole, very few categories).
#  - Subplots: Multiple related plots for comparison.
#  - Color Maps: Show extra data dimension (size, value, density, etc.).
# ============================================================================
```
