```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
import pandas as pd
import numpy as np
```

```python
# Create hierarchical company data
company_data = pd.DataFrame({
    'Region': ['North', 'North', 'South', 'South', 'East', 'West'],
    'Dept': ['Sales', 'Marketing', 'Sales', 'Operations', 'Tech', 'Tech'],
    'Team': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Employees': [50, 30, 40, 35, 60, 25],
    'Satisfaction': [3.8, 3.2, 4.1, 3.7, 4.5, 4.0]
})
```

```python
company_data
```

```python
fig = px.treemap(company_data,
                 path=['Region', 'Dept', 'Team'],
                 values='Employees',
                 color='Satisfaction',
                 color_continuous_scale='RdBu',
                 title='Treemap: Employees and Satisfaction by Region/Dept/Team')
fig.show()
```

```python
prod_sales = pd.DataFrame({
    'Line': ['Electronics', 'Electronics', 'Clothing', 'Clothing'],
    'Product': ['Phones', 'Laptops', 'Jeans', 'Shirts'],
    'Market': ['Online', 'Retail', 'Online', 'Retail'],
    'Units': [1200, 900, 1900, 1600],
    'Growth': [8, 10, 15, 12]
})
```

```python
fig = px.sunburst(prod_sales,
                  path=['Line', 'Product', 'Market'],
                  values='Units',
                  color='Growth',
                  color_continuous_scale='Tealrose',
                  title='Sunburst: Units Sold by Line/Product/Market')
# When/Why: Best for circular, nested category comparisons with proportional representation
fig.show()
```

```python
map_data = pd.DataFrame({
    'City': ['CityA', 'CityB', 'CityC', 'CityD'],
    'Lat': [28.7, 19.1, 13.1, 22.6],
    'Lon': [77.1, 72.8, 80.2, 88.4],
    'Customers': [2900, 210, 2600, 180]
})
```

```python
fig = px.scatter_geo(map_data,
                     lat="Lat", lon="Lon",
                     hover_name="City", size="Customers",
                     projection="natural earth",
                     title="Bubble Map: Customers by City (Fictional Data)")
# When/Why: Visualize real-world locations with variable-specific marker sizes
fig.show()
```

```python
np.random.seed(42)
customer_data = pd.DataFrame({
    'Age': np.random.randint(22, 60, 40),
    'Income': np.random.randint(25000, 150000, 40),
    'Score': np.random.rand(40) * 100,
    'Group': np.random.choice(['A', 'B', 'C'], 40)
})
```

```python
fig = px.scatter_3d(customer_data,
                    x='Age', y='Income', z='Score',
                    color='Group',
                    title="3D Scatter: Customer Segmentation")
# When/Why: Great to visualize clustering or separation based on three continuous variables
fig.show()
```

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
# Example function: two-lobed "hill" surface
Z = np.exp(-(X**2 + Y**2)) - np.exp(-((X-1.5)**2 + (Y-1.5)**2))

fig = go.Figure(go.Contour(
    z=Z, x=x, y=y, colorscale='Magma',
    contours_coloring='lines', line_width=2))
fig.update_layout(title="Contour: Custom Surface Levels (Synthetic Data)")
# When/Why: Find peaks/valleys, hotspots, or critical lines in a 2D field
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Example dataset
tips = sns.load_dataset('tips')

# Scatter plot using Seaborn
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
# Bar plot using Seaborn
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum)
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

```python
# Demo dataset
df = px.data.tips()

# Bar chart
fig = px.bar(df, x='day', y='total_bill', color='sex', barmode='group',
             title='Total Bill by Day and Gender')
fig.show()
```

```python
# Pie chart
fig = px.pie(df, values='tip', names='day', title='Distribution of Tips by Day')
fig.show()
```

```python
# Demo dataset
df = px.data.stocks()

# Line plot
fig = px.line(df, x='date', y='GOOG', title='Google Stock Prices over Time')
fig.show()
```

```python
# Demo dataset
df = px.data.tips()

# Box plot
fig = px.box(df, x='day', y='total_bill', color='smoker',
             title='Box Plot of Total Bill by Day and Smoker Status')
fig.show()
```
