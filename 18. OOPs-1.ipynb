```python
class person12:

    dob = 1999

    def __init__ (self, name, age, email):
        self.name = name       
        self.age = age        
        self.email = email 

    def age1(self, current_year):
        return current_year- self.dob
        
```

```python
var3 = person12("rahul1", 29, "rahul1@email.com")
```

```python
var3.age1(2025)
```

```python
class person13:

    __current_year = 2025

    def __init__ (self, name, age, email):
        self.name = name       
        self.age = age        
        self.email = email 

    def age1(self, DOB):
        return DOB - self.__current_year
```

```python
var4 = person13("rahul1", 29, "rahul1@email.com")
```

```python
var4.age
```

```python
var4.age1(1995)
```
