```python
# Parent class
class Vehicle:
    def general_info(self):
        print("Vehicles are used for transportation.")
```

```python
# Child class 1  
class Car(Vehicle):
    def general_info(self):
        super().general_info()
        print("car class--- Vehicles are used for transportation.")
    
    def car_info(self):
        print("Car has 4 wheels.")
        super().general_info()
```

```python
obj = Vehicle()
```

```python
obj.general_info()
```

```python
obj1 = Car()
```

```python
obj1.general_info()
```

```python
obj1.car_info()
```

```python
class ProtectedExample:
    def __init__(self):
        self._protected_var = "I'm protected"  # Protected attribute
    
    def _protected_method(self):  # Protected method
        return "This is a protected method"
    
    def access_protected(self):
        """Public method to access protected members"""
        return self._protected_var
```

```python
obj1 = ProtectedExample()
```

```python
obj1._protected_method()
```

```python
class PrivateExample:
    def __init__(self):
        self.__private_var = "I'm private"  # Private attribute
        self.public_var = "I'm public"
    
    def __private_method(self):  # Private method
        return "This is a private method"
    
    def access_private(self):
        """Public method to access private members"""
        return self.__private_var + " - " + self.__private_method()
```

```python
obj3 = PrivateExample()
```

```python
obj3._PrivateExample__private_method()
```

```python
class MathUtils:
   
    def add(self, a, b):  # Static method
        """Adds two numbers"""
        return a + b
    
    @staticmethod
    def is_even(num):  # Static method
        """Checks if number is even"""
        return num % 2 == 0
    
    @staticmethod
    def factorial(n):  # Static method
        """Calculates factorial"""
        if n <= 1:
            return 1
        return n * MathUtils.factorial(n - 1)
```

```python
obj4 = MathUtils()
```

```python
obj4.add(3,8)
```

```python
obj4.factorial(5)
```

```python
obj4.is_even(567)
```
