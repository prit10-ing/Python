```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df[df['Age'] > 30]
```

```python
df[(df['Age'] > 25) & (df['City'] == 'Paris')]
```

```python
df[(df['Age'] > 25) | (df['City'] == 'Paris')]
```

```python
# Using isin() method
df[df['City'].isin(["London"])]
```

```python
# String operations in boolean indexing
print(df[df['Name'].str.contains('A')])    # Names containing 'A'
```

```python
print(df[df['Name'].str.startswith('B')])  # Names starting with 'A'
```
