### Multiple inheritence

```python
class bank :
    def transaction(self) :
        print("total transaction value ")
    def account_opening(self):
        print("this will show you your account open status ")
    def deposite(self):
        print("this will show your deposited amount")
    def test(self):                          # override code
        print("this is a test methond form bank class ")


class HDFC_bank:
    def hdfc_to_icici(self):
        print("this will show you all the transaction happend to icici throuhg hdfc")
    def test(self):           #override code
        print("this is a test method form hdfc bank ")

class ineron_bank:
    def account_status_icici(self):
        print("print a account status in icici ")

class icici(HDFC_bank, bank, ineron_bank):      
    pass
```

```python
obj = bank()
```

```python
obj1 = HDFC_bank()
```

```python
obj2 = ineron_bank()
```

```python
obj3 = icici()
```

### Hirerchical inheritence

```python
# Parent class
class Vehicle:
    def start(self):
        print("Vehicle started")

# Child class 1
class Car(Vehicle):
    def drive(self):
        print("Driving a car")

# Child class 2
class Bike(Vehicle):
    def ride(self):
        print("Riding a bike")
```

```python
# Creating objects of child classes
c = Car()
b = Bike()

# Access parent method from both child classes
c.start()  # Inherited from Vehicle
c.drive()  # Car-specific method

b.start()  # Inherited from Vehicle
b.ride()   # Bike-specific method
```

### hybrid inheritence

```python
# Parent class
class Vehicle:
    def general_info(self):
        print("Vehicles are used for transportation.")

# Child class 1 (inherits from Vehicle)
class Car(Vehicle):
    def car_info(self):
        print("Car has 4 wheels.")

# Child class 2 (inherits from Vehicle)
class Bike(Vehicle):
    def bike_info(self):
        print("Bike has 2 wheels.")

# Grandchild class (inherits from both Car and Bike)
class Electricvehicle(Car, Bike):
    def electric_info(self):
        print("Electric bike runs on battery.")
```

```python
obj = Electricvehicle()
```

```python
obj.general_info()
```

### Encapsulation

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited {amount}. New balance: {self.__balance}"
        return "Invalid deposit amount"

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return f"Withdrew {amount}. New balance: {self.__balance}"
        return "Invalid withdrawal amount"

    def get_balance(self):
        return self.__balance

```

```python
# Testing the class
account = BankAccount("Alice", 1000)
print(account.deposit(500))  # Deposited 500. New balance: 1500
print(account.withdraw(200))  # Withdrew 200. New balance: 1300
print(account.get_balance())  # 1300
# print(account.__balance)  # This would raise an AttributeError
```

### polymorphism

```python
class Animal:
    def make_sound(self):
        pass  # This is a placeholder; subclasses will override it

class Dog(Animal):
    def make_sound(self):
        return "Woof!"

class Cat(Animal):
    def make_sound(self):
        return "Meow!"

class Bird(Animal):
    def make_sound(self):
        return "Chirp!"

```

```python
o1= Dog()
o2 = Cat()
o3 = Bird()
```

```python
# Testing polymorphism
animals = [o1,o2,o3]
animals
```

```python
for animal in animals:
    print(animal.make_sound())
```
