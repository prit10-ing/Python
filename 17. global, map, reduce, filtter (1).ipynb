### map

```python
# Simple example - convert numbers to their cubes
numbers = [1, 2, 3, 4, 5]
```

```python
list(map(lambda x: x**3, numbers))
```

```python
# Using lambda with map to find cubes
cubes = list(map(lambda x: x**3, numbers))
print(f"Original numbers: {numbers}")
print(f"Cubes: {cubes}")
```

```python
tup = (3,4,5)
```

```python
list(map(lambda v: v**2, tup))
```

### filter

```python
# Example - find numbers divisible by 5
numbers = [10, 13, 15, 20, 23, 25, 30, 33, 35]
```

```python
list(map(lambda x: x % 5 == 0, numbers))
```

```python
10 % 5 == 0
```

```python
list(filter(lambda x: x % 5 == 0, numbers))
```

```python
# Using lambda with filter
divisible_by_5 = list(filter(lambda x: x % 5 == 0, numbers))
print(f"Original numbers: {numbers}")
print(f"Divisible by 5: {divisible_by_5}")
```

### Reduce

```python
from functools import reduce

# Example - find product of all numbers
numbers = [2, 3, 4, 5]
```

```python
reduce(lambda x, y: x * y, numbers)
```

```python
# Using lambda with reduce
product = reduce(lambda x, y: x * y, numbers)
print(f"Numbers: {numbers}")
print(f"Product: {product}")
```

```python
celsius = [0, 10, 20, 30]
fahrenheit = list(map(lambda x: x * 9/5 + 32, celsius))
print(f"Fahrenheit: {fahrenheit}") 
```

```python
words = ["apple", "banana", "kiwi"]
lengths = list(map(lambda x: len(x), words))
print(f"Word lengths: {lengths}") 
```

```python
numbers = [1, 2, 3, 4, 5, 6, 7]
evens = list(filter(lambda x: x % 2 != 0, numbers))
print(f"Even numbers: {evens}") 
```

```python
words = ["apple", "banana", "kiwi", "cherry"]
long_words = list(filter(lambda x: len(x) > 5, words))
print(f"Long words: {long_words}")
```

```python
numbers = [4, 7, 1, 3]
total = reduce(lambda x, y: x / y, numbers)
print(f"Sum: {total}") 
```

```python
def decorator_function(original_function):
    def wrapper_function():
        print("Wrapper: Before the original function")
        original_function()
        print("Wrapper: After the original function")
    return wrapper_function
```

```python
def my_decorator(func):
    def wrapper():
        print("Before function is called")
        func()
        print("After function is called")
    return wrapper
```

```python
@my_decorator  # This is equivalent to say_hello = my_decorator(say_hello)
def say_hello():
    print("Hello!")

say_hello()
```

```python
@decorator_function
def say_hello():
    print("Hello!")

say_hello()
```

```python
@decorator_function
def mufun11():
    var1 = "java"
    print(var1)
```

```python
mufun11()
```
